### Investigating DF of Unique Businesses

```python
unique_biz = df.drop_duplicates(subset=['ABI'])
unique_biz.head()
```

```python
unique_biz.groupby('STATE').count()['Unnamed: 0.1'].sort_values(ascending=False)
#States with the most plants (100+) = Arkansas, Georgia, Texas, Alabama, North Carolina, California
```

```python
unique_biz['EMPLOYEE SIZE (5) - LOCATION'].sort_values(ascending=False)
```
