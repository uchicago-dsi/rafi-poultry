## Taking post-processed and filtered model output and mapping using folium

```python
import folium
import json
```

```python
with open('../data/filtered_5_tifs.geojson') as f:
    five_data = json.load(f)
```

```python
m = folium.Map(location=[34.7837, -78.5769], zoom_start=10)
```

```python
folium.GeoJson(
    five_data
).add_to(m)
```
