## Taking post-processed model output and mapping using folium

```python
import folium
import json
```

```python
with open('../data/m_3407819_sw_17_1_20120510_processed.geojson') as f:
    geojson_data = json.load(f)
```

```python
DUB_LAT = 34.6568
DUB_LNG = -78.7264
```

```python
m = folium.Map(location=[DUB_LAT, DUB_LNG], zoom_start=12)
```

```python
m
```

```python
folium.GeoJson(
    geojson_data
).add_to(m)
```
