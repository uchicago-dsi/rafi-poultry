```python
# Find the latest naics as the only naics for the company
def get_naics(df): 
    NAICS = []
    for i in range(len(df)):
        row = df.iloc[i,1:].to_list()
        l = 0
        while l in range(len(row)):
            if pd.isna(row[l]) == False:
                NAICS.append(row[l])
                break
            else:
                l += 1
        continue
    return NAICS           
naics['NAICS'] = get_naics(naics) # The number of NAICS shows that each company has a NAICS now
# Merge companies and naics and filter to left only auction house
companies_naics = companies.merge(naics[['DunsNumber','NAICS']], how='inner', on='DunsNumber')
companies_naics = companies_naics[companies_naics['NAICS'] == 424520.0]
companies_naics
```
