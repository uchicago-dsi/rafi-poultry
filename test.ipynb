### actual important stuff

```python
df = pd.DataFrame(np.nan, index=list(range(len(chesapeake_gdf))), columns=['state', 'address', 'company', 'plant_access'])
```

```python
df['geometry'] = chesapeake_gdf['geometry']
```

```python
df['geometry'] = df['geometry'].apply(lambda x: x.centroid)
```

```python
df
```

```python
formatted_gdf = gpd.GeoDataFrame(
        {
            "state": df['state'],
            "address": df['address'],
            "company": df['company'],
            "plant_access": df['plant_access'],
            "geometry": df['geometry']
        }
    )
```

```python
formatted_gdf.to_file("data/msft/chesapeake_bay.geojson", driver="GeoJSON")
```
